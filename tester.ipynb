```python
pth_for_charades_ds = '/workspace/h5/i3d_finetuned'
pth_for_charades_h5 = '/workspace/h5/i3d-charades.hdf5'
```

```python
import numpy as np
import os
```

```python
tmp = np.load(os.path.join(pth_for_charades_ds, '0A8CF.npy'))
```

```python
tmp.squeeze().shape
```

```python
f2 = h5py.File(pth_for_charades_h5, 'w')
```

```python
all_dir_charades = os.listdir(pth_for_charades_ds)
```

```python
from tqdm import tqdm
for npf in tqdm(all_dir_charades):
    fulldir = os.path.join(pth_for_charades_ds, npf)
    npf_ = np.load(fulldir)
    npf_ = npf_.squeeze()
    f2.create_dataset(npf[:-4]+'/i3d_features', data=npf_, dtype='f8')
```

```python
f2.close()
```
